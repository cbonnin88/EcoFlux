```python
import pandas as pd
```

# **Loading the raw event data**

```python
df_amplitude = pd.read_csv('ecoflux_amplitude_events.csv')
```

# **Defining the Product Logic**

```python
# Mapping the activity_score to the events that I want to track in Amplitude
```

```python
def map_events(score):
  if score > 0.85:
    return 'Schedule Smart Charge'
  elif score > 0.30:
    return 'App Open'
  else:
    return 'Background Session'
```

```python
df_amplitude['event_type'] = df_amplitude['activity_score'].apply(map_events)
```

# **Create Event Properties**

```python
df_amplitude['carbon_saved_kg'] = df_amplitude.apply(
    lambda row: round(row['activity_score'] * 12.5,2)
    if row['event_type'] == 'Schedule Smart Charge' else 0, axis=1
)
```

# **Filter for 'Active' Events**

```python
amplitude_df = df_amplitude[df_amplitude['event_type'] != 'Background Session'].copy()
```

# **Selecting and Renaming Columns for Amplitude**

```python
final_columns = [
    'user_id',
    'event_type',
    'time',
    'country',
    'subscription_plan',
    'device_type',
    'carbon_saved_kg'
]

amplitude_final = amplitude_df[final_columns]
```

```python
display(amplitude_final.head())
```

```python
amplitude_final.to_csv('ecoflux_amplitude_final.csv', index=False)

print(f'Success ! {len(amplitude_final)} events prepared for Amplitude upload')
```
